```python
# implement stack using python list
class Stack:
    def __init__(self):
        self.stack = []
    def push(self, item):
        self.stack.append(item)
    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return None
    def is_empty(self):
        return len(self.stack) == 0
    def size(self):
        return len(self.stack)
# --- IGNORE ---
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)   
print(stack.pop())  # Output: 3
print(stack.peek()) # Output: 2
```

```python
# valid parentheses leetcode 20
def isValid(s):
    stack = []
    mapping = {")": "(", "}": "{", "]": "["}
    for char in s:
        if char in mapping:
            print(char)
            top_element = stack.pop() if stack else '#'
            print(mapping[char], top_element
            )
            if mapping[char] != top_element:
                return False
        else:
            stack.append(char)
    return not stack
print(isValid("()"))        # True
print(isValid("()[]{"))    # True
```

```python
# min stack leetcode 155
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []
    def push(self, val: int) -> None:
        self.stack.append(val)
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)
    def pop(self) -> None:
        if self.stack:
            val = self.stack.pop()
            if val == self.min_stack[-1]:
                self.min_stack.pop()
    def top(self) -> int:
        if self.stack:
            return self.stack[-1]
        return None
    def getMin(self) -> int:
        if self.min_stack:
            return self.min_stack[-1]
        return None
min_stack = MinStack()
min_stack.push(-2)
min_stack.push(0)
min_stack.push(-3)
print(min_stack.getMin())  # Output: -3
min_stack.pop()
print(min_stack.top())      # Output: 0
print(min_stack.getMin())   # Output: -2
```

```python
 # max stack leetcode 716 - review later
class MaxStack:
    def __init__(self):
        self.stack = []
        self.max_stack = []
    def push(self, x: int) -> None:
        self.stack.append(x)
        if not self.max_stack or x >= self.max_stack[-1]:
            self.max_stack.append(x)
    def pop(self) -> int:
        if self.stack:
            val = self.stack.pop()
            if val == self.max_stack[-1]:
                self.max_stack.pop()
            return val
        return None
    def top(self) -> int:
        if self.stack:
            return self.stack[-1]
        return None
    def peekMax(self) -> int:
        if self.max_stack:
            return self.max_stack[-1]
        return None
    def popMax(self) -> int:
        if not self.max_stack:
            return None
        max_val = self.max_stack.pop()
        buffer = []
        while self.stack and self.stack[-1] != max_val:
            buffer.append(self.stack.pop())
        if self.stack:
            self.stack.pop()  # Remove the max element
        while buffer:
            self.push(buffer.pop())
        return max_val
```

```python
# Daily temperatures leetcode 739
def dailyTemperatures(T):
    result = [0] * len(T)
    stack = []
    for i in range(len(T)-1, -1, -1):
        while stack and T[i] >= T[stack[-1]]:
            stack.pop()
        if stack:
            result[i] = stack[-1] - i
        stack.append(i)
    return result   
print(dailyTemperatures([73,74,75,71,69,72,76,73]))  # Output: [1,1,4,2,1,1,0,0]
```
