```python
# count vowel permutation leetcode 1220
def countVowelPermutation(n):
    MOD = 10**9 + 7
    dp = [1] * 5  # a, e, i, o, u

    for _ in range(1, n):
        a = (dp[1] + dp[2] + dp[4]) % MOD  # e, i, u -> a
        e = (dp[0] + dp[2]) % MOD          # a, i -> e
        i = (dp[1] + dp[3]) % MOD          # e, o -> i
        o = dp[2] % MOD                     # i -> o
        u = (dp[2] + dp[3]) % MOD          # i, o -> u
        dp = [a, e, i, o, u]

    return sum(dp) % MOD

print(countVowelPermutation(1))  # Output: 5
print(countVowelPermutation(2))  # Output: 10
```

```python
# ones and zeroes leetcode 474
def findMaxForm(strs, m, n):
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for s in strs:
        zeros = s.count('0')
        ones = s.count('1')

        for i in range(m, zeros - 1, -1):
            for j in range(n, ones - 1, -1):
                dp[i][j] = max(dp[i][j], dp[i - zeros][j - ones] + 1)

    return dp[m][n]
```
